### 1. What is the key factor that makes the difference between ideas that can, and cannot be examined and tested statistically? What would you describe is the key "criteria" defining what a good null hypothesis is? And what is the difference between a null hypothesis and an alternative hypothesis in the context of hypothesis testing?

Statistical testing is dependent on a couple of factors. First, we should be able to get data and it should be quantitative. That is to say, we should be able to get numbers out of the data that we put in. Next, it should be able to be replicated- meaning, whatever experiment is done should be able to be enacted multiple times with many samples. There should also be an element of randomness for the data and between each sample, so that there is some element of probability that can be analyzed.

As for a good null hypothesis, it should fit the following criteria- it should be testable (meaning, it fits the definition of statistical testing listed above), it should be specific with regards to detail so that cases that match or stray from the null prediction are easy to tell and unambiguous, and it should also be falsifiable- meaning, with the right evidence, we should be able to prove the null hypothesis wrong.

Moreover, the difference between a null hypothesis ($H_0$) and an alternative hypothesis ($H_1$ or $H_a\!$) is as follows: <br>
The null hypothesis represents the default hypothesis of "no effect"- meaning, that the difference seen can be explained by random variability rather than due to a factor that we can meaningfully attribute to. However, the alternative hypothesis represents the claim that is to be tested- that there is a measurable effect which can contradict the null hypothesis.

### 3. The second "Pre-lecture" video (above) explains that we "imagine a world where the null hypothesis is true" when calculating a p-value? Explain why this is in your own words in a way that makes the most sense to you.

The reason why we assume that the null hypothesis is true is because we are trying to answer the question of *"if nothing special is going on, how likely is it for this to be happening?"*, which was a very good quote given by ChatGPT and can be found in the logs at the top of this document. The P-value represents how likely it is for us to see data which looks like the one we received for even more extreme than the data we have. It is a measure of how likely it is for our data to turn out this way. In this way, we can quite easily tell whether we believe the null hypothesis to be true. If the explanation is easily explainable and the data does not look too out of the ordinary, then it is reasonable to conclude that the null hypothesis could be true. Conversely, if the data which we see is very unlikely to occur naturally, it tells us that something strange is happening and that our null hypothesis is likely inaccurate.

In order to explain the ideas of p-value and the null hypothesis, it is best to imagine an example which looks at a coin flip. Let's pretend that a friend gives you a coin and tells you that it's fair, so you flip it 100 times to test it and it lands on heads 99 times and tails only once. In this case, we will assume that our null hypothesis (meaning, the value we would *expect* to see if everything was normal and the coin really was fair) is that the coin should land on heads approximately 50 times and the same for tails. In our case, it is *possible* for the coin to be entirely fair and land on heads 99 times. However, we would say that it is extremely rare for it to happen. This is exactly what the p-value represent- the likelihood of an event happening. The higher the p-value, the more likely it is for an event to happen. In this case, because of the likelihood to see statistics as extreme or even moreso given the null hypothesis, we can conclude that it does indeed make the null hypothesis look ridiculous- sure, it's possible that your friend has a knack for flipping coins to always land on one side, but it's much easier to just accept the proposition that maybe the coin just isn't fair- and this answer is much easier to accept due to the much higher likelihood (and p-value) of the alternative hypothesis occuring.

### 5. Gntrkn (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a **p-value** using a "50/50 coin-flipping" model for the assumption of the **null hypothesis** $H_0$ that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against $H_0$. 

![Rodin's sculpture, "The Kiss"
](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d1/Rodin_-_Le_Baiser_06.jpg/409px-Rodin_-_Le_Baiser_06.jpg)

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

```python
import numpy as np
import random
import plotly.express as px

# Define the parameters
n_couples = 124   # Total number of couples
observed_right_tilts = 80  # Observed number of couples tilting right
n_simulations = 10000  # Number of simulations

# Simulation function to simulate the number of right-tilts (like coin flips)
def simulate_p_value(n_couples, n_simulations, observed_right_tilts):
    simulated_right_tilts = []

    # Run the simulation n_simulations times
    for _ in range(n_simulations):
        # Simulate 124 coin flips (50/50 probability)
        right_tilts = sum([random.choice([True, False]) for _ in range(n_couples)])
        simulated_right_tilts.append(right_tilts)

    # Convert to a NumPy array for easier calculation
    simulated_right_tilts = np.array(simulated_right_tilts)

    # Calculate the p-value (how often we got 80 or more right tilts)
    p_value = np.sum(simulated_right_tilts >= observed_right_tilts) / n_simulations

    return p_value, simulated_right_tilts

# Run the simulation
p_value, simulated_right_tilts = simulate_p_value(n_couples, n_simulations, observed_right_tilts)

# Print the p-value
print(f"P-value: {p_value}")

# Plot a histogram of the simulated right-tilts
fig = px.histogram(simulated_right_tilts, nbins=50, title='Simulated Right-Tilts Distribution',
                   labels={'value': 'Number of Right Tilts'},
                   template='plotly_dark')
fig.add_vline(x=observed_right_tilts, line_dash="dash", line_color="red", annotation_text="Observed Right Tilts", annotation_position="top right")

# Show the plot
fig.show()
```

With such low P value, there is very strong evidence against the null hypothesis.

Work with a ChatBot to adjust the code from "Demo II of the Week 5 TUT" (which revisits the "Vaccine Data Analysis Assignment" from Week 04 HW "Question 8") in order to compute a p-value for a "one sided" (or "one tailed") hypothesis test rather than the "two sided" (or "two tailed") version it provides. Describe (perhaps with the help of your ChatBot) what changed in the code; how this changes the interpretation of the hypothesis test; and whether or not we should indeed expect the p-value to be smaller in the "one tailed" versus "two tailed" analysis. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('data.csv')

# Calculate the change in health score (FinalHealthScore - InitialHealthScore)
data['HealthScoreChange'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Function to perform bootstrapping
def bootstrap(data, num_samples=1000, stat_function=np.mean):
    """
    data: Array-like data to bootstrap.
    num_samples: Number of bootstrap samples to draw.
    stat_function: The function used to calculate the statistic (e.g., np.mean for mean).
    """
    n = len(data)
    bootstrapped_stats = []
    
    # Generate bootstrap samples and compute the statistic for each sample
    for _ in range(num_samples):
        # Randomly sample the data with replacement
        bootstrap_sample = np.random.choice(data, size=n, replace=True)
        
        # Calculate the statistic and store it
        stat = stat_function(bootstrap_sample)
        bootstrapped_stats.append(stat)
    
    return np.array(bootstrapped_stats)

# Apply bootstrapping to the HealthScoreChange column
num_bootstrap_samples = 10000
bootstrap_means = bootstrap(data['HealthScoreChange'], num_samples=num_bootstrap_samples)

# Hypothesized mean change (usually 0 if you're testing whether there's an increase)
observed_mean_change = np.mean(data['HealthScoreChange'])

# Calculate the proportion of bootstrap samples greater than or equal to the observed mean change
p_value = np.sum(bootstrap_means >= observed_mean_change) / num_bootstrap_samples

# Print the one-sided p-value
print(f"One-sided p-value: {p_value:.4f}")

# Visualize the bootstrap distribution
plt.figure(figsize=(8, 6))
plt.hist(bootstrap_means, bins=30, edgecolor='black', color='skyblue')
plt.axvline(np.mean(bootstrap_means), color='blue', linestyle='-', label=f'Mean: {np.mean(bootstrap_means):.2f}')
plt.axvline(observed_mean_change, color='red', linestyle='--', label=f'Observed Mean Change: {observed_mean_change:.2f}')
plt.title('Bootstrapped Means of Health Score Change (One-Tailed Test)')
plt.xlabel('Mean Change in Health Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

In the code above, we introduced a one sided p-value in which we only evaluated those with health changes greater than 0 and did not evaluate those with health changes at or below 0. We expect to see that the p-value will be smaller than original. When we consider this problem and the original experiment devised by Fisher and Bristol, there are many differences to be pondered. First, we must consider the elephant in the room of sample size- that is to say, in the original experiment, there is a sample of one individual drinking from one cup of tea. In this hypothetical experiment, we aim to evaluate the ability of a population to correctly identify the order of tea and milk being poured into a cup. While this hypothetical lacks detail, we can assume that each student drinks from two cups only and must identify one with milk poured first and one with the milk poured after. In this case, we can construct a null hypothesis and an alternative hypothesis that is fitting for this experiment.<br><br>

**Null hypothesis and alternative hypothesis:**<br>
In this case, our null hypothesis is that the order of the tea and milk being poured has no effect and that they do not effect the taste. We will set this to be our null hypothesis, which should also produce a mean of the null hypothesis- which we will assume is the value of 40, since if there really is no difference in the taste, then by randomly choosing we should expect approximately half of the people to guess the right answer by random. This should intuitively make sense because if there really is no difference, then the choice becomes as random as a coin toss with both tasting the same, unless there is some extra psychology at play.

Next, below is the code to simulate this experiment and bootstrapping the dataset.

```python
# code goes here
import numpy as np
import random
import plotly.express as px

# The list below represents the raw data of those who can and cannot teste the
# difference between milk or tea first. We will use True for those that can
# taste the difference, and False for those who cannot.
data = []

# This simple for loop creates the default data.
for i in range(80):
    if i < 50:
        data.append(True)
    else:
        data.append(False)


def bootstrap(data: [str], n: int = 100000) -> np.ndarray:
    bootstrapped_means = []

    for i in range(n):
        temp_total = 0
        for i in range(len(data)):
            if random.choice(data):
                temp_total += 1

        bootstrapped_means.append(temp_total)

    return np.array(bootstrapped_means)


# Run the bootstrap function to generate bootstrapped means
bootstrapped_means = bootstrap(data)

lower = np.percentile(bootstrapped_means, 2.5)
upper = np.percentile(bootstrapped_means, 97.5)

# Create a histogram using plotly.express
fig = px.histogram(bootstrapped_means, nbins=50, title='Histogram of Bootstrapped Means',
                   labels={'value': 'Number of People Who Can Taste the Difference'},
                   template='plotly_dark') # Thank you ChatGPT for the dark mode graph (that was its only input)

# Add the lines for the upper and lower percentiles
fig.add_vline(x=lower)
fig.add_vline(x=upper)

# Show the histogram
fig.show(renderer='png')
```

Above we have some code which helps visualize the bootstrapped dataset, courtesy of not-so-much ChatGPT (except, it did pick this cool looking colour scheme so I gotta give it credit for that) with some simple and intuitive python code. The lines at the 2.5 and 97.5 percentiles mark our 95% confidence interval- meaning, for all our statistical trials, we expect to see results fall between these values approximately 95% of the time. This isn't to say that the values outside of the confidence interval is impossible, but we will say that they are less likely to occur. In this simple experiment, since the null hypothesis of 40 falls outside the confidence interval, we can reject it as the best explanation for this dataset is better explained by the alternative hypothesis that perhaps there really is some effect from putting the milk before the tea.<br><br>
